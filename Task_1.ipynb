```python
df.columns=['Date', 'Time', 'Booking_ID', 'Booking_Status', 'Customer_ID',
       'Vehicle_Type', 'Pickup_Location', 'Drop_Location', 'Avg_VTAT',
       'Avg_CTAT', 'Cancelled_Rides_by_Customer',
       'Reason_for_cancelling_by_Customer', 'Cancelled_Rides_by_Driver',
       'Driver_Cancellation_Reason', 'Incomplete_Rides',
       'Incomplete_Rides_Reason', 'Booking_Value', 'Ride_Distance',
       'Driver_Ratings', 'Customer_Rating', 'Payment_Method']
```

```python
df.head()
```

```python
df['Cancelled_Rides_by_Customer'].fillna(0,inplace=True)
df['Cancelled_Rides_by_Driver'].fillna(0,inplace=True)
df['Reason_for_cancelling_by_Customer'].fillna('Not Applicable', inplace=True)
df['Driver_Cancellation_Reason'].fillna('Not Applicable', inplace=True)
df['Incomplete_Rides'].fillna(0, inplace=True)
df['Incomplete_Rides_Reason'].fillna('Not Applicable', inplace=True)
df['Avg_VTAT'].fillna(0, inplace=True)
df['Avg_CTAT'].fillna(0, inplace=True)
df['Ride_Distance'].fillna(0, inplace=True)
df['Driver_Ratings'].fillna(0, inplace=True)
df['Customer_Rating'].fillna(0, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
```

```python
df['Booking_Status'].value_counts()
```

```python
df['Vehicle_Type'].value_counts().plot(kind='bar')
```

```python
df['Payment_Method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

```python
df[['Cancelled_Rides_by_Customer','Cancelled_Rides_by_Driver']].sum().plot(kind='bar')
```

```python
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
df['Hour'].value_counts().sort_index().plot(kind='bar')
```

```python
total_revenue = df[df['Booking_Status'] == 'Completed']['Booking_Value'].sum()
print("Total Revenue: ", total_revenue)
```

```python
avg_revenue = df[df['Booking_Status'] == 'Completed']['Booking_Value'].mean()
print("Average Revenue", avg_revenue)
```

```python
total_distance = df.groupby('Vehicle_Type')['Ride_Distance'].sum()
plt.figure(figsize=(7,5))
total_distance.plot(kind='bar', color=['red','blue','purple','brown','green','black','maroon'])
plt.title("Total Distance Covered by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Total Distance (km)")
plt.show()
```

```python
revenue_by_vehicle = df[df['Booking_Status'] == 'Completed'].groupby('Vehicle_Type')['Booking_Value'].sum()
plt.figure(figsize=(7,5))
revenue_by_vehicle.plot(kind='bar', color=['skyblue','orange','green','red','yellow','purple','pink'])
plt.title("Revenue by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
status_counts = df['Booking_Status'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(status_counts, labels=status_counts.index,
        autopct='%1.1f%%', startangle=90, colors=['lightgreen','salmon','skyblue','cyan','lavender'])
plt.title("Booking Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df['Driver_Ratings'], color='skyblue', edgecolor='black')
plt.title("Distribution of Driver Ratings")
plt.xlabel("Driver Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df['Customer_Rating'], color='orange', edgecolor='black')
plt.title("Distribution of Customer Ratings")
plt.xlabel("Customer Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
avg_ratings = df.groupby('Vehicle_Type')[['Driver_Ratings','Customer_Rating']].mean()
avg_ratings.plot(kind='bar', figsize=(8,5))
plt.title("Average Ratings by Vehicle Type")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```
