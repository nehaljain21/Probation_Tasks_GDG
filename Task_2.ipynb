**Handling Missing Values**

```python
for col in ['Teacher_Quality','Parental_Education_Level','Distance_from_Home']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

**EDA and Encoding**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))
plt.scatter(df['Hours_Studied'], df['Exam_Score'], alpha=0.5, color='blue')
plt.title('Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
```

```python
avg_scores_by_school = df.groupby('School_Type')['Exam_Score'].mean()
print(avg_scores_by_school)
plt.figure(figsize=(7,5))
avg_scores_by_school.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Exam Score by School Type')
plt.xlabel('School Type')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
avg_scores_by_gender = df.groupby('Gender')['Exam_Score'].mean()
print(avg_scores_by_gender)
plt.figure(figsize=(7,5))
avg_scores_by_gender.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Exam Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Exam Score')
plt.show()
```

```python
print(df["School_Type"].unique())
print(df["Peer_Influence"].unique())
print(df["Parental_Education_Level"].unique())
print(df["Distance_from_Home"].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in ['Extracurricular_Activities','School_Type','Internet_Access','Learning_Disabilities','Gender']:
    df[col] = le.fit_transform(df[col])
```

```python
ordinal_map = {'Low':1, 'Medium':2, 'High':3}
for col in ['Parental_Involvement','Access_to_Resources','Motivation_Level','Family_Income','Teacher_Quality']:
  df[col]=df[col].map(ordinal_map)
df['Peer_Influence'] = df['Peer_Influence'].map({"Negative":1,"Neutral":2,"Positive":3})
df['Parental_Education_Level'] = df['Parental_Education_Level'].map({"High School":1,"College":2,"Postgraduate":3})
df['Distance_from_Home'] = df['Distance_from_Home'].map({"Near":1,"Moderate":2,"Far":3})
```

```python
df.head()
```

```python
import seaborn as sns
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
print(df["Attendance"].min())
print(df["Attendance"].max())
```

```python
bins = [0, 60, 70, 80, 90, 100]
labels = ['0-60','61-70', '71-80', '81-90', '91-100']
df['Attendance_Range'] = pd.cut(df['Attendance'], bins=bins, labels=labels, right=True)
grouped = df.groupby('Attendance_Range',observed=True)['Exam_Score'].mean()
print(grouped)
plt.figure(figsize=(7,5))
grouped.plot(kind='bar', color='skyblue')
plt.title('Average Exam Score by Attendance Range')
plt.xlabel('Attendance Range')
plt.ylabel('Average Exam Score')
plt.show()
```

**Training and testing model**

```python
x = df.drop(['Exam_Score','Attendance_Range'], axis=1)
y = df['Exam_Score']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

**Predictions**

```python
y_pred = lr.predict(x_test)
```

**Evaluation**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", mse**0.5)
print("R² Score:", r2)
```

```python
df_prediction=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
plt.figure(figsize=(6,5))
corr = df_prediction.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap: Actual vs Predicted")
plt.show()
```

**Testing on new record**

```python
students_list = []
while True:
    choice = input("Add student's details (Y/N)? = ")
    if choice in "Yy":
        print('\nEnter the following details of the student: (Enter in CAPITALS)\n')
        student = {'Hours_Studied': int(input("Hours Studied = ")),
            'Attendance': int(input("Attendance (1-100) = ")),
            'Parental_Involvement': input("Parental Involvement (LOW/MEDIUM/HIGH) = "),
            'Access_to_Resources': input("Access to Resources (LOW/MEDIUM/HIGH) = "),
            'Extracurricular_Activities': input("Extracurricular Activities (YES/NO) = "),
            'Sleep_Hours': int(input("Sleep Hours = ")),
            'Previous_Scores': int(input("Previous Scores = ")),
            'Motivation_Level': input("Motivation Level (LOW/MEDIUM/HIGH) = "),
            'Internet_Access': input("Internet Access (YES/NO) = "),
            'Tutoring_Sessions': int(input("Tutoring Sessions = ")),
            'Family_Income': input("Family Income (LOW/MEDIUM/HIGH) = "),
            'Teacher_Quality': input("Teacher Quality (LOW/MEDIUM/HIGH) = "),
            'School_Type': input("School Type (PUBLIC/PRIVATE) = "),
            'Peer_Influence': input("Peer Influence (NEGATIVE/NEUTRAL/POSITIVE) = "),
            'Physical_Activity': int(input("Physical Activity = ")),
            'Learning_Disabilities': input("Learning Disabilities (YES/NO) = "),
            'Parental_Education_Level': input("Parental Education Level (HIGH SCHOOL/COLLEGE/POST GRADUATE) = "),
            'Distance_from_Home': input("Distance from Home (NEAR/MODERATE/FAR) = "),
            'Gender': input("Gender (MALE/FEMALE) = ")}
        students_list.append(student)
    elif choice in "Nn":
        break
    else:
        print("Invalid Input")
new_student = pd.DataFrame(students_list)
```

```python
new_student.head()
```

```python
for col in ['Extracurricular_Activities','School_Type','Internet_Access','Learning_Disabilities','Gender']:
    new_student[col] = le.fit_transform(new_student[col])
ordinal_map1 = {'LOW':1, 'MEDIUM':2, 'HIGH':3}
for col in ['Parental_Involvement','Access_to_Resources','Motivation_Level','Family_Income','Teacher_Quality']:
  new_student[col]=new_student[col].map(ordinal_map1)
new_student['Peer_Influence'] = new_student['Peer_Influence'].map({"NEGATIVE":1,"NEUTRAL":2,"POSITIVE":3})
new_student['Parental_Education_Level'] = new_student['Parental_Education_Level'].map({"HIGH SCHOOL":1,"COLLEGE":2,"POST GRADUATE":3})
new_student['Distance_from_Home'] = new_student['Distance_from_Home'].map({"NEAR":1,"MODERATE":2,"FAR":3})
```

```python
predicted_scores = lr.predict(new_student)
```
